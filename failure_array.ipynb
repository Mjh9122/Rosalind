```python
from Dna import Dna
import numpy as np
from tqdm import tqdm
```

```python
strings = []

with open('Inputs/rosalind_kmp(3).txt') as f:
    dna = Dna(f.readlines())
    
s = dna.string
s[-10:]
```

```python
P = np.zeros((len(s)), int)

for i in tqdm(range(1, len(s))):
    last_value = P[i-1]
    if s[last_value] == s[i]:
        P[i] = last_value + 1
    else:
        for j in range(last_value, 0, -1):
            if s[:j] == s[i - j + 1: i + 1]:
                P[i] = j
                break

out = ''
mine = ''.join([f'{int(i)} ' for i in P])
print(mine)
```
