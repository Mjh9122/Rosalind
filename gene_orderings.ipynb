```python
import itertools
```

```python
def orderings(n):
    ret = []
    perms = itertools.permutations(range(1, n + 1))
    for perm in perms:
        for i in range(n + 1):
            combinations = itertools.combinations(perm, i)
            for comb in combinations:
                out = ''
                for num in perm:
                    if num in comb:
                        out += f'-{num} '
                    else:
                        out += f'{num} '
                ret.append(out)
    return ret
```

```python
ret = orderings(5)
print(len(ret))
for r in ret:
    print(r)
```
