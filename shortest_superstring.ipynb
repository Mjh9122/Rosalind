```python
from Dna import Dna
import itertools
from tqdm import tqdm
```

```python
strings = []

with open('Inputs/rosalind_long(1).txt') as f:
    lines = f.readlines()[1:]
    active_str = ''
    for l in lines:
        if l[0] == '>':
            strings.append(active_str)
            active_str = ''
        else:
            active_str += l.strip()
```

```python
def find_overlaps(strings):
    max_overlap = (0, strings[0], strings[1])
    for s1, s2 in itertools.permutations(strings, 2):
        for i in range(1, min(len(s1), len(s2))):
            if s1[:i] == s2[-i:]:
                if i > max_overlap[0]:
                    max_overlap = (i, s1, s2)
    return max_overlap
    
```

```python
for i in tqdm(range(len(strings)-1)):
    overlap, s1, s2 = find_overlaps(strings)
    s3 = s2[:-overlap] + s1
    strings.remove(s1)
    strings.remove(s2)
    strings.append(s3)

print(strings)
```
