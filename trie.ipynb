```python
import numpy as np
import networkx as nx
```

```python
g = nx.DiGraph()
g.add_node((1, 'root'))
with open('./Inputs/rosalind_trie.txt') as f:
    lines = f.readlines()
    strings = [l.strip() for l in lines]

strings
```

```python
count = 2
current = (1, 'root')

for string in strings:
    current = (1, 'root')
    is_prefix = True
    for letter in string:
        n_ints = [n[0] for n in nx.neighbors(g, current)]
        n_letters = [n[1] for n in nx.neighbors(g, current)]
        if letter in n_letters:
            current = (n_ints[n_letters.index(letter)], letter)
        else:
            g.add_node((count, letter))
            g.add_edge(current, (count, letter))
            current = (count, letter)
            count += 1


#nx.draw_kamada_kawai(g, with_labels = True)
```

```python
with open('./Output/out.txt', 'w') as f:
    for u, v in sorted(list(g.edges())):
        f.write(f'{u[0]} {v[0]} {v[1]}\n')
```
